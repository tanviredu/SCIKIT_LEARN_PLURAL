```
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('housing.csv')
```

```
df = df.dropna()
```

```
df.shape
```

```
## removing the uppr cap data
## the cluster data that congesed in the top
## other wise it will make a skew the data
df.loc[df['median_house_value'] == 500001].count()
```

```
### whn make linear regression
### rule out the cluster data
### because it does not fit the linear regression
df = df.drop(df.loc[df['median_house_value'] == 500001].index)
```

```
df.head()
```

```
## now we can apply the liner regression
## because in previous some data was cluster in the top
df['ocean_proximity'].unique()
```

```
## use the one hot encoding
target_column = ['ocean_proximity']
df = pd.get_dummies(df,columns=target_column)
```

```
df.head()
```

```
df.shape
```

```
df.sample(10)
```

```
X = df.drop('median_house_value',axis=1)
Y = df[['median_house_value']]
```

```
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.2)
```

```
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression(normalize=True).fit(x_train,y_train)
```

```
linear_model
```

```
print("Train Score  : {}".format(linear_model.score(x_train,y_train)))
```

```
print("Test Score  : {}".format(linear_model.score(x_test,y_test)))
```

```
## finding the coefficient
predictors = x_train.columns
```

```
print(predictors)
print(linear_model.coef_[0])
```

```
## when we make the linear model
## the equation is y=mx+c
## in this case there is more than 2 coefficient
## lets print the coefficient use for rach category
coeff = pd.Series(linear_model.coef_[0],predictors).sort_values()
```

```
coeff
```

```
# these are the coefficient used in making the regression line
plt.plot(coeff)
```
